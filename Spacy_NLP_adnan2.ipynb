```python

outputs = []
counts = []

for txt in pravda_df.iloc[:,3]:
    txt = txt.replace('"','')
    doc = nlp(txt)
    
    entities = []
    
    for ent in doc.ents:
        if ent.label_ == "GPE" or ent.label_ == "LOC":
            entities.append(ent.text)
            
    # Join the entities into a string
    output = ", ".join(entities)
    
    # Append the output to the list of outputs
    outputs.append(output)

    ### Count the number of locations in outputs
    location_count = Counter(outputs)
    
    ## append the outputs to the list of counts
    counts.append(location_count)

    
    

# Add a new column named the output
pravda_df['output'] = outputs

# add a new column named the output
pravda_df['count'] = counts


print(pravda_df)
```

```python
output_1s = []  # GPE
output_2s = []  # LOC

for txt in pravda_df.iloc[:,3]:
    txt = txt.replace('"','')
    doc = nlp(txt)
    
    entities = []
    entity = []
    
    for ent in doc.ents:
        if ent.label_ == "GPE":
            entities.append(ent.text)
            
    # Join the entities into a string
    output_1 = ", ".join(entities)
    
    # Append the output to the list of outputs
    output_1s.append(output_1)

    for ent in doc.ents:
        if ent.label_ == "LOC":
            entity.append(ent.text)
            
    output_2 = ", ".join(entity)
    
    output_2s.append(output_2)
    
    
    
# Add a new column named the output

pravda_df['output_1'] = output_1s
pravda_df['output_2'] = output_2s
print(pravda_df)
```

```python
import spacy
from collections import Counter
import en_core_web_sm
nlp = en_core_web_sm.load()
%matplotlib inline
```

```python
pravda_df = pd.read_excel("march10_pravda.xlsx")
pravda_df.head(10)
```

```python
tokens = nlp(''.join(str(pravda_df.text.tolist())))
```

```python
items = [x.text for x in tokens.ents]
Counter(items).most_common(20)
```

```python
location_list = []
for ent in tokens.ents:
    if ent.label_ == 'GPE' or ent.label == 'LOC':
        location_list.append(ent.text)
        
location_counts = Counter(location_list).most_common(20)
df_location = pd.DataFrame(location_counts, columns =['text', 'count'])
```
