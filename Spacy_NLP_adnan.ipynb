```python
import spacy
from collections import Counter
import en_core_web_sm
nlp = en_core_web_sm.load()
%matplotlib inline
```

```python
pravda_df = pd.read_excel("march10_pravda.xlsx")
pravda_df.head(10)
```

```python
tokens = nlp(''.join(str(pravda_df.text.tolist())))
```

```python
items = [x.text for x in tokens.ents]
Counter(items).most_common(20)
```

```python
location_list = []
for ent in tokens.ents:
    if ent.label_ == 'GPE':
        location_list.append(ent.text)
        
location_counts = Counter(location_list).most_common(20)
df_location = pd.DataFrame(location_counts, columns =['text', 'count'])
```
